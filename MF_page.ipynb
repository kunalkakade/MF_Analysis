```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel("MF_Analysis.xlsx")
df["Crisil Rank"] = df["Crisil Rank"].replace('-',0)
df
```

```python
CN = df.groupby(["Category Name","Crisil Rank"])
```

```python
CN.get_group(('Sectoral/Thematic', 4))["AuM (Cr)"].describe()
```

```python
ST_5 = CN.get_group(('Sectoral/Thematic', 5))[["Scheme Name" ,"AuM (Cr)"]]
ST_4 = CN.get_group(('Sectoral/Thematic', 4))[["Scheme Name" ,"AuM (Cr)"]]
ST_3 = CN.get_group(('Sectoral/Thematic', 3))[["Scheme Name" ,"AuM (Cr)"]]
```

```python
ST_543= pd.concat([ST_3, ST_4, ST_5],axis=0)
```

```python
my_plot = ST_543.plot(kind='bar',x="Scheme Name",title="AuM" ,figsize=(15,15))
```
